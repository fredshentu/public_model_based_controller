```python
import matplotlib.pyplot as plt
import numpy as np
import os
import csv

%matplotlib inline
```

```python
DATA_LOCATION = "/Users/dianchen/rllab/data/local/"

def plot_rewards(ax, exp_name, run_names):
    lines = []
    for i, dirname in enumerate(os.listdir(DATA_LOCATION + exp_name)):
        if dirname in run_names:
            rewards = []
            with open(DATA_LOCATION + exp_name + "/" + dirname + "/progress.csv") as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    rewards.append(row["AverageReturn"])
            line, = ax.plot(rewards)
            lines.append(line)
    plt.title("Rewards v.s Episode")
    plt.legend(lines, run_names, loc="best")
            
def get_max_rewards(exp_name):
    max_rewards = []
    for i, dirname in enumerate(os.listdir(DATA_LOCATION + exp_name)):
        rewards = []
        with open(DATA_LOCATION + exp_name + "/" + dirname + "/progress.csv") as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                rewards.append(float(row["AverageReturn"]))
        max_rewards.append(np.max(rewards))
#     print ("Max reward for %s: %0.3f" % (exp_name, np.mean(max_rewards)))
    print ("Max reward for %s: %0.3f" % (exp_name, np.max(max_rewards)))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,6))
plot_rewards(ax1, "Box3DReach-DDPG", [
    "Box3DReach-DDPG_2017_04_22_20_13_49_0001",
    "Box3DReach-DDPG_2017_04_22_20_13_49_0002",
    "Box3DReach-DDPG_2017_04_22_20_13_49_0003",
])
plot_rewards(ax2, "Box3D-Reach-DDPG-ICM", [
    "Box3D-Reach-DDPG-ICM_2017_04_23_00_21_50_0001",
    "Box3D-Reach-DDPG-ICM_2017_04_23_00_21_50_0002",
    "Box3D-Reach-DDPG-ICM_2017_04_23_00_21_50_0003",
])
plot_rewards(ax2, "Box3D-Reach-DDPG-ICM-20", [
    "Box3D-Reach-DDPG-ICM-20_2017_04_23_03_09_22_0001",
    "Box3D-Reach-DDPG-ICM-20_2017_04_23_03_09_22_0002",
    "Box3D-Reach-DDPG-ICM-20_2017_04_23_03_09_22_0003",
])

plot_rewards(ax2, "Box3D-Reach-DDPG-ICM", [
    "Box3D-Reach-DDPG-ICM-20_2017_04_23_03_09_22_0001",
    "Box3D-Reach-DDPG-ICM-20_2017_04_23_03_09_22_0002",
    "Box3D-Reach-DDPG-ICM-20_2017_04_23_03_09_22_0003",
])
```
